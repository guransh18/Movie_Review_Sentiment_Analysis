```python
import streamlit as st
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the saved model and TF-IDF vectorizer
model = pickle.load(open('sentiment_model.pkl', 'rb'))
tfidf = pickle.load(open('tfidf_vectorizer.pkl', 'rb'))

# Page setup
st.set_page_config(page_title="Movie Review Sentiment Analyzer", layout="wide")

# App header
st.title("🎬 Movie Review Sentiment Analyzer")
st.markdown("""
Welcome to the Sentiment Analysis App! This tool uses a Machine Learning model to predict the sentiment (positive or negative) of movie reviews.
""")

# Sidebar Information
st.sidebar.title("About the Project")
st.sidebar.info(
    "This sentiment analysis model was trained on movie reviews. "
    "It processes reviews using TF-IDF vectorization and predicts sentiments using a machine learning classifier."
)

# Function to predict sentiment
def predict_sentiment(review):
    processed_review = tfidf.transform([review])
    prediction = model.predict(processed_review)
    return 'Positive 😊' if prediction[0] == 1 else 'Negative 😞'

# Input box for user reviews
st.header("📝 Enter Your Movie Review")
user_review = st.text_area("Type your movie review below:")

if st.button("Predict Sentiment"):
    if user_review:
        sentiment = predict_sentiment(user_review)
        st.success(f"The predicted sentiment is: **{sentiment}**")
    else:
        st.warning("Please enter a review to analyze.")

# Sentiment Distribution Visualization (Demo Data)
st.header("📊 Sentiment Distribution (Sample Data)")
sample_data = {
    "Review": [
        "The movie was fantastic and I loved every second of it.",
        "Terrible movie. I wasted my time.",
        "The movie is not bad, quite enjoyable.",
        "I wouldn't recommend this movie to anyone.",
        "Just okay, nothing special.",
        "Brilliant direction and acting!",
        "Not good at all, very disappointing.",
        "Amazing visuals but weak story.",
        "Worst movie I've ever seen.",
        "The plot was not interesting, but stunning visuals."
    ],
    "Sentiment": [1, 0, 1, 0, 0, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(sample_data)

# Plotting sentiment distribution
sentiment_counts = df['Sentiment'].value_counts().sort_index()
plt.figure(figsize=(6, 4))
plt.bar(['Negative', 'Positive'], sentiment_counts, color=['#FF6347', '#90EE90'])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
st.pyplot(plt)

# Generate WordCloud from positive and negative reviews
st.header("☁️ WordCloud of Reviews")

positive_reviews = ' '.join(df[df['Sentiment'] == 1]['Review'])
negative_reviews = ' '.join(df[df['Sentiment'] == 0]['Review'])

col1, col2 = st.columns(2)

with col1:
    st.subheader("Positive Reviews")
    pos_wordcloud = WordCloud(background_color='white', width=400, height=300).generate(positive_reviews)
    plt.figure(figsize=(5, 3))
    plt.imshow(pos_wordcloud, interpolation='bilinear')
    plt.axis('off')
    st.pyplot(plt)

with col2:
    st.subheader("Negative Reviews")
    neg_wordcloud = WordCloud(background_color='white', width=400, height=300).generate(negative_reviews)
    plt.figure(figsize=(5, 3))
    plt.imshow(neg_wordcloud, interpolation='bilinear')
    plt.axis('off')
    st.pyplot(plt)

# Footer
st.markdown("---")
st.write("Developed by Guransh Singh 🚀 | Powered by Streamlit and Machine Learning")
```
